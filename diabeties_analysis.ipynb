```python
import pandas as pd
from sqlalchemy import create_engine

# 1. CSV file path
csv_path = "diabetes.csv"   # full path   safe

# 2. Load CSV into DataFrame
df = pd.read_csv(csv_path)

print("CSV loaded successfully")
print(df.head())

# 3. PostgreSQL connection details
username = "postgres"
password = "newpassword"
host = "localhost"
port = "5432"
database = "diabetes_project"

# 4. Create SQLAlchemy engine
engine = create_engine(
    f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}"
)

# 5. Send DataFrame to PostgreSQL
df.to_sql(
    name="diabetes_data",
    con=engine,
    if_exists="replace",   # table auto-create 
    index=False
)

print("Data successfully imported into PostgreSQL")
```
